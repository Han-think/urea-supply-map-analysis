# Urea Supply Map Analysis

A compact, GitHub-safe analysis of the 2021 urea inventory sample, restricted to Gyeonggi-do and Incheon because the observed station sample is heavily concentrated in those two regions.

## Scope

- The station sample is treated as an observed sample, not a national census.
- Station names, street addresses, and phone numbers are excluded from the public dataset.
- Demand estimates use explicit scenario assumptions and public aggregated vehicle/mileage tables.

```
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path('..')
DATA = ROOT / 'data' / 'processed'
FIG = ROOT / 'figures'

region = pd.read_csv(DATA / 'region_supply_balance.csv')
municipality = pd.read_csv(DATA / 'municipality_demand_supply.csv')
station = pd.read_csv(DATA / 'station_priority_anonymized.csv')
allocation = pd.read_csv(DATA / 'allocation_simulation_1million_l.csv')
region
```

```
cols = ['province', 'annual_demand_base_l', 'observed_inventory_l', 'inventory_cover_days', 'safety_stock_gap_l']
region[cols].sort_values('annual_demand_base_l', ascending=False)
```

```
municipality[['province','municipality','annual_demand_base_l','observed_inventory_l','safety_stock_gap_l']].head(15)
```

```
fig, ax = plt.subplots(figsize=(8, 4.5))
region.plot(x='province', y=['annual_demand_low_l','annual_demand_base_l','annual_demand_high_l'], kind='bar', ax=ax)
ax.set_title('Annual Urea Demand Scenarios')
ax.set_ylabel('Liters per year')
ax.set_xlabel('')
plt.tight_layout()
```

```
top = municipality.nlargest(15, 'annual_demand_base_l').sort_values('annual_demand_base_l')
fig, ax = plt.subplots(figsize=(8, 6))
ax.barh(top['municipality'], top['annual_demand_base_l'], color='#de2d26')
ax.set_title('Top 15 Municipalities by Estimated Demand')
ax.set_xlabel('Annual demand, liters')
plt.tight_layout()
```

```
fig, ax = plt.subplots(figsize=(7,5))
scatter = ax.scatter(station['longitude'], station['latitude'], s=(station['inventory_l'].clip(lower=0)/station['inventory_l'].max()*220+20), c=station['final_priority_score'], cmap='Reds', alpha=0.72, edgecolor='white', linewidth=0.4)
ax.set_title('Observed Station Priority Sample')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.colorbar(scatter, ax=ax, label='Priority score')
plt.tight_layout()
```
